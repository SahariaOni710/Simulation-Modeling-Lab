```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Observed frequencies
observed = np.array([45, 35, 25])

# Expected frequencies for a uniform distribution
expected = np.ones_like(observed) * np.mean(observed)  # Equal expected frequencies
```

```python
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)
```

```python
print("Observed:", observed)
print("Expected:", expected)
print(f"Chi-square Statistic = {chi2_stat:.3f}")
print(f"P-value = {p_value:.3f}")

# Conclusion
if p_value < 0.05:
    print("Reject the null hypothesis â†’ data does not fit a uniform distribution.")
else:
    print("Fail to reject the null hypothesis â†’ data fits a uniform distribution.")
```

```python
categories = ['A', 'B', 'C']
plt.bar(categories, observed, alpha=0.7, label='Observed')
plt.bar(categories, expected, alpha=0.7, label='Expected')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.title('Observed vs Expected Frequencies (Uniform Test)')
plt.legend()
plt.show()
```
