```python
import random  # to create random numbers

def estimate_pi(num_points, radius):
    inside = 0  # how many points inside the circle

    for _ in range(num_points):
        x = random.uniform(0, radius)  # pick a random number from 0 to radius
        y = random.uniform(0, radius)
        if x*x + y*y <= radius*radius:
            inside += 1  # it's inside the circle

    pi = 4 * inside / num_points
    print("Estimated pi for", num_points, "points is:", pi)

# Your details
R = 20
radius = R / 10  #2
num_points = 200 * 10  # 10 letters in "NUSRAT SAHARIA"

# Run the estimation
estimate_pi(num_points, radius)
```
