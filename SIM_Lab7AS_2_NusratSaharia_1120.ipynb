```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
observed2 = np.array([50, 30, 15, 5])
expected2 = np.array([40, 30, 20, 10])
```

```python
chi2_stat2, p_value2 = stats.chisquare(f_obs=observed2, f_exp=expected2)
print("Observed:", observed2)
print("Expected:", expected2)
print(f"Chi-square Statistic = {chi2_stat2:.3f}")
print(f"P-value = {p_value2:.3f}")

if p_value2 < 0.05:
    print("Reject the null hypothesis â†’ observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis â†’ observed data fits the expected distribution.")
```

```python
categories2 = ['A', 'B', 'C', 'D']
plt.bar(categories2, observed2, alpha=0.7, label='Observed')
plt.bar(categories2, expected2, alpha=0.7, label='Expected')
plt.xlabel('Grades')
plt.ylabel('Frequency')
plt.title('Observed vs Expected Frequencies (Custom Test)')
plt.legend()
plt.show()
```
